**Data Reducction**

```python
#Removing the Max Cardinal Direction
data=data.drop(["Max Cardinal Direction"], axis=1)
data.head()
```

```python
data['Daily Mean Windspeed']=data['Daily Mean Windspeed'].fillna(data['Daily Mean Windspeed'].mean())
print("Mean of Daily Mean Windspeed is:", data['Daily Mean Windspeed'].mean())
print("Mean of Daily Mean Windspeed is:", data['Daily Mean Windspeed'].median())
print("Mean of Daily Mean Windspeed is:", data['Daily Mean Windspeed'].mode())
```

Fill all null in numeric data

```python
numeric_cols = data.select_dtypes(include=['number']).columns
data[numeric_cols] = data[numeric_cols].apply(lambda x: x.fillna(x.mean()))
```

```python
# Generate the correlation matrix
data['Daily Total Rainfall'] = pd.to_numeric(data['Daily Total Rainfall'], errors='coerce')
data['Daily Mean Windspeed (Beaufort)'] = pd.to_numeric(data['Daily Mean Windspeed (Beaufort)'], errors='coerce')

numerical_data = data.select_dtypes(include=['number'])


correlation_matrix = numerical_data.corr()


print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
